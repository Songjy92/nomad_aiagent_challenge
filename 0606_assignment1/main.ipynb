```python
import openai

client = openai.OpenAI()

url = "https://nomad-movies-2.nomadcoders.workers.dev/"

PROMPT = """
I have a following functions in my api:

base_url = "https://nomad-movies-2.nomadcoders.workers.dev/"

function 1. `get_popular_movies()` : returns a list of popular movies in /movies
function 2. `get_movie_details(id)` : return details of a movie in /movies/:id
fucntion 3. `get_movie_credits(id)` : return credits, including cast and crew of a movie in /movies/:id/credits

Please answer with the name of the function and the arguments if needed.
For example, if I ask "What are the popular movies?", you can answer like this:
get_popular_movies()

"""
def get_ai_response(user_input):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content":PROMPT},
            {"role": "user", "content": user_input},
        ],
    )
    return response.choices[0].message.content
```

```python
q1 = get_ai_response("지금 인기 있는 영화가 무엇인지 알려줘")
print(q1)
```

```python
q2 = get_ai_response("movie ID 550에 해당하는 영화가 무엇인지 알려줘")
print(q2)
```

```python
q3 = get_ai_response("movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘")
print(q3)
```
